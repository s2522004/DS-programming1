## 基本のリクエスト

```python
# アクセスしたいWebサイトのURLを指定
url = 'https://www.musashino-u.ac.jp'

# WebサイトにアクセスしてHTML情報を取得
# レスポンスを変数に格納
res = requests.get(url)
```

## パラメータ付きのリクエスト

```python
# url = "https://www.musashino-u.ac.jp/search.html?query=データサイエンス"
url = "https://www.musashino-u.ac.jp/search.html"
params = {
    'query': 'データサイエンス'
}
res = requests.get(url, headers=headers, params=params)
res.encoding = res.apparent_encoding  # レスポンスのエンコーディングを設定
print(f"ステータスコード：{res.status_code}")  # ステータスコードを表示
print(f"レスポンスボディ:{res.text}")  # レスポンスボディを表示
```

## BeautifulSoupでHTMLソースから任意の情報を取得する

```python
url = 'https://www.musashino-u.ac.jp'
res = requests.get(url)
res.encoding = res.apparent_encoding  # レスポンスのエンコーディングを設定
print(f"ステータスコード：{res.status_code}")  # ステータスコードを表示
```

```python
# HTMLのソースコードを解析
soup = BeautifulSoup(res.text, 'html.parser')
print(f"soupの型：{type(soup)}")  # soupの型を表示
```

```python
# ソースコード全体から<a>タグを抽出
soup_a = soup.a
print(f"<a>タグの内容：{soup_a}")  # <a>タグの内容を表示
print(f"<a>タグの型：{type(soup_a)}")  # <a>タグの型を表示

#  ソースコード全体から<p>タグを抽出
soup_p = soup.p
print(f"<p>タグの内容：{soup_p}")  # <p>タグの内容を表示
print(f"<p>タグの型：{type(soup_p)}")  # <p>タグの型を表示
```

```python
# ソースコード全体から<div>タグを抽出
soup_div = soup.div
# print(f"<div>タグの内容：{soup_div}")  # <div>タグの内容を表示
# print(f"<div>タグの型：{type(soup_div)}")  # <div>タグの型を表示

soup_div_div = soup_div.div
print(f"<div>タグの中の<div>タグの内容：{soup_div_div}")  # <div>タグの中の<div>タグの内容を表示
print(f"<div>タグの中の<div>タグの型：{type(soup_div_div)}")  # <div>タグの中の<div>タグの型を表示
```

```python
# BeautifulSoupオブジェクトとTagオブジェクトは、"."でつなげていくことで、HTML要素を階層的にたどることができます。
soup_div_div = soup.div.div
print(f"<div>タグの中の<div>タグの内容：{soup_div_div}")  # <div>タグの中の<div>タグの内容を表示
print(f"<div>タグの中の<div>タグの型：{type(soup_div_div)}")  # <div>タグの中の<div>タグの型を表示
```

```python
# Tagオブジェクト.stringで、タグの中のテキスト情報を取得できる。
print(f"<a>タグのテキスト内容：{soup_a.string}")  # <a>タグのテキスト内容を表示

# コンテンツが無い場合、Noneが返される。
print(f"<div>タグの中の<div>タグのテキスト内容：{soup_div_div.string}")  # <div>タグの中の<div>タグのテキスト内容を表示
```

## 指定したHTMLタグで要素を検索

```python
# 指定したタグが使われている要素のうち、最初に見つかった要素を取得する。
h2 = soup.find('h2')
print(f"{h2}")  # 最初に見つかった<h2>タグを表示
print(f"変数の型：{type(h2)}")  # <h2>タグの型を表示
```

```python
h2_all = soup.find_all('h2')
# print(f"<h2>タグのリスト：{h2_all}")  # 見つかったすべての<h2>タグを表示
# print(f"型：{type(h2_all)}")  # 型を表示

for h2 in h2_all:
    if h2.span is not None:
        print(f"{h2.span.string}")  # 各<h2>タグの中の<span>タグのテキスト内容を表示
```

```python
# a_list = soup.find_all('a')
# for a in a_list:
#    print(f"{a.string}")  # 各<a>タグのテキスト内容を表示

# 検索結果の数に制限をかける
a_list_limited = soup.find_all('a', limit=5)
for a in a_list_limited:
    print(f"{a.string}")  # 各<a>タグのテキスト内容を表示
```

```python
# 複数のタグを指定して検索することも可能
# 例えば、<h2>タグと<h3>タグの両方を取得したい場合は、次のようにします。
h2_h3_list = soup.find_all(['h2', 'h3'], limit=5)
for tag in h2_h3_list:
    print(f"{tag.string}")  # 各タグのテキスト内容を表示
```

## 下記のページから出演映画のタイトルを取得する。
https://abehiroshi.la.coocan.jp/movie/eiga.htm 


```python
! pip install requests beautifulsoup4
```

```python
import requests
from bs4 import BeautifulSoup
import re

# URL指定
url = "https://abehiroshi.la.coocan.jp/movie/eiga.htm"

# HTML取得
response = requests.get(url)
response.encoding = response.apparent_encoding  # 日本語文字化け対策
html = response.text

# BeautifulSoup解析
soup = BeautifulSoup(html, "html.parser")

movies = []

# <a>タグなどにタイトルが含まれている
for tag in soup.find_all(["a", "b", "font", "td"]):
    text = tag.get_text(strip=True)
    # 映画タイトルの形式（例：「ドラゴン桜」など日本語っぽい行を抽出）
    if text and len(text) > 1:
        # 不要な文字列を除外
        if not any(x in text for x in ["阿部寛", "（", "）", "★", "◇", "公式サイト", "映画"]):
            # ひらがな・カタカナ・漢字を含む行を抽出
            if re.search(r'[ぁ-んァ-ン一-龥]', text):
                movies.append(text)

# 重複削除
movies = list(dict.fromkeys(movies))

print("=== 阿部寛 出演映画リスト ===")
for i, title in enumerate(movies, start=1):
    print(f"{i}. {title}")

print(f"\n合計: {len(movies)}作品")
```
